```python
import numpy as np
import matplotlib.pyplot as plt


# Define the function
def f(x, y):
    return (x - 2) ** 2 + (y - 3) ** 2 + np.sin(3 * x) + np.cos(3 * y)


# Create a grid of (x, y) points
x = np.linspace(-1, 5, 100)
y = np.linspace(0, 6, 100)
X, Y = np.meshgrid(x, y)

# Calculate Z values
Z = f(X, Y)

# Create the 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection="3d")
surf = ax.plot_surface(X, Y, Z, cmap="viridis", edgecolor="none")

# Add labels and title
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y) - Loss")
ax.set_title("Loss Surface Plot")
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

# Optimization Landscape Visualization

## 3D Loss Surface Plot

This visualization demonstrates the **loss landscape** that optimization algorithms must navigate during neural network training.

### The Loss Function

**Mathematical form:** $f(x, y) = (x-2)^2 + (y-3)^2 + \sin(3x) + \cos(3y)$

This function combines:
- **Quadratic terms:** $(x-2)^2 + (y-3)^2$ creates a smooth bowl shape
- **Trigonometric terms:** $\sin(3x) + \cos(3y)$ adds local minima and maxima

### Key Concepts Illustrated

1. **Global Minimum:** The lowest point on the entire surface (around x=2, y=3)
2. **Local Minima:** Smaller valleys that can trap gradient descent
3. **Saddle Points:** Regions where the surface curves up in one direction and down in another
4. **Non-convexity:** Multiple valleys and peaks make optimization challenging

### Why This Matters for Neural Networks

Real neural network loss surfaces are:
- **High-dimensional** (thousands to billions of parameters)
- **Highly non-convex** (many local minima)
- Difficult to visualize directly

This 2D example helps build intuition for why:
- **Learning rate** matters (large steps might overshoot the minimum)
- **Initialization** is important (starting point affects which minimum we find)
- **Advanced optimizers** (Adam, RMSprop) can navigate complex landscapes better than vanilla gradient descent
