## Part 1: Activation Functions

Activation functions introduce non-linearity into neural networks, enabling them to learn complex patterns.

### Sigmoid Activation Function

**Formula:** $\sigma(x) = \frac{1}{1 + e^{-x}}$

**Properties:**
- Output range: (0, 1)
- S-shaped curve
- Useful for: Binary classification (output layer)
- **Drawback:** Vanishing gradients for large |x| values

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define Tanh function
def tanh(x):
    return np.tanh(x)

# 2. Create data
x = np.linspace(-5, 5, 200)
y = tanh(x)

# 3. Plot
plt.figure(figsize=(6, 4))
plt.plot(x, y, label='Tanh', color='orange')
plt.title('Tanh Activation Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# 4. Save
plt.savefig('tanh_plot.png')
plt.close()

print("Tanh plot saved as tanh_plot.png")
```

### Tanh (Hyperbolic Tangent) Activation Function

**Formula:** $\tanh(x) = \frac{e^{2x} - 1}{e^{2x} + 1}$

**Properties:**
- Output range: (-1, 1)
- Zero-centered (unlike sigmoid)
- Useful for: Hidden layers in traditional networks
- **Advantage over sigmoid:** Zero-centered outputs help with gradient flow

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define ReLU function
def relu(x):
    return np.maximum(0, x)

# 2. Create data
x = np.linspace(-5, 5, 200)
y = relu(x)

# 3. Plot
plt.figure(figsize=(6, 4))
plt.plot(x, y, label='ReLU', color='green')
plt.title('ReLU Activation Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# 4. Save
plt.savefig('relu_plot.png')
plt.close()

print("ReLU plot saved as relu_plot.png")
```

### ReLU (Rectified Linear Unit) Activation Function

**Formula:** $\text{ReLU}(x) = \max(0, x)$

**Properties:**
- Output range: [0, ∞)
- Simple and computationally efficient
- **Most popular for deep networks**
- **Advantages:** Solves vanishing gradient problem, faster training
- **Drawback:** "Dying ReLU" problem (neurons can get stuck at 0)

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define Leaky ReLU function
def leaky_relu(x, alpha=0.1):
    return np.maximum(alpha * x, x)

# 2. Create data
x = np.linspace(-5, 5, 200)
y = leaky_relu(x, alpha=0.1) # Using 0.1 alpha to make the leak visible

# 3. Plot
plt.figure(figsize=(6, 4))
plt.plot(x, y, label='Leaky ReLU (α=0.1)', color='red')
plt.title('Leaky ReLU Activation Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# 4. Save
plt.savefig('leaky_relu_plot.png')
plt.close()

print("Leaky ReLU plot saved as leaky_relu_plot.png")
```

### Leaky ReLU Activation Function

**Formula:** $\text{Leaky ReLU}(x) = \max(\alpha x, x)$ where $\alpha$ is typically 0.01 or 0.1

**Properties:**
- Output range: (-∞, ∞)
- Small negative slope for x < 0
- **Advantage over ReLU:** Prevents dying ReLU problem by allowing small negative gradients
- Common in modern deep networks

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define SiLU function
def silu(x):
    return x * (1 / (1 + np.exp(-x)))

# 2. Create data
x = np.linspace(-5, 5, 200)
y = silu(x)

# 3. Plot
plt.figure(figsize=(6, 4))
plt.plot(x, y, label='SiLU (Swish)', color='purple')
plt.title('SiLU (Swish) Activation Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# 4. Save
plt.savefig('silu_plot.png')
plt.close()

print("SiLU plot saved as silu_plot.png")
```

### SiLU (Sigmoid Linear Unit) / Swish Activation Function

**Formula:** $\text{SiLU}(x) = x \cdot \sigma(x) = \frac{x}{1 + e^{-x}}$

**Properties:**
- Output range: (~-0.28, ∞)
- Smooth, non-monotonic function
- Self-gated (uses its own value for gating)
- **Used in:** Modern architectures like EfficientNet, Transformers
- **Advantage:** Smooth gradient flow, often outperforms ReLU in deep networks

```python
import matplotlib.pyplot as plt
import numpy as np

# Set a random seed for reproducibility
np.random.seed(42)

# 1. Generate Original Data
# Create data with different means and different standard deviations
mean = [3, 5]
cov = [[1, 0.5], [0.5, 4]]  # std_x=1, std_y=2
n_samples = 300
original_data = np.random.multivariate_normal(mean, cov, n_samples)

# 2. Create Zero-Centered Data
data_mean = np.mean(original_data, axis=0)
centered_data = original_data - data_mean

# 3. Create Normalized Data (Standardized)
data_std = np.std(centered_data, axis=0)
normalized_data = centered_data / data_std

# 4. Plot all three steps
plt.figure(figsize=(18, 6))

# Plot 1: Original Data
plt.subplot(1, 3, 1)
plt.scatter(original_data[:, 0], original_data[:, 1], c='red', alpha=0.5)
plt.title('Original Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xlim(-2, 8)
plt.ylim(-2, 12)
plt.gca().set_aspect('equal', adjustable='box')

# Plot 2: Zero-Centered Data (Same 'tall' shape, new center)
plt.subplot(1, 3, 2)
plt.scatter(centered_data[:, 0], centered_data[:, 1], c='green', alpha=0.5)
plt.title('Zero-Centered Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xlim(-7, 7)  # Note the new axis limits
plt.ylim(-7, 7)
plt.gca().set_aspect('equal', adjustable='box')

# Plot 3: Normalized Data (New 'squished' shape, same center)
plt.subplot(1, 3, 3)
plt.scatter(normalized_data[:, 0], normalized_data[:, 1], c='purple', alpha=0.5)
plt.title('Normalized Data (Standardized)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xlim(-7, 7)  # Same limits as plot 2 to show the scaling
plt.ylim(-7, 7)
plt.gca().set_aspect('equal', adjustable='box')

plt.tight_layout()
plt.savefig('preprocessing_steps_chart_FIXED.png')
```

## Part 2: Data Preprocessing Pipeline

Proper data preprocessing is crucial for neural network training success.

### Visualization: Three-Step Preprocessing

This demonstrates the standard preprocessing pipeline with synthetic 2D data:

1. **Original Data:** Raw data with arbitrary mean and varying scales
   - Features may have different ranges
   - Not centered at origin

2. **Zero-Centered Data:** Subtract the mean from each feature
   - Centers the data around (0, 0)
   - **Purpose:** Helps gradient descent converge faster
   - Shape/spread remains unchanged

3. **Normalized Data (Standardized):** Divide by standard deviation
   - Each feature now has mean=0 and std=1
   - **Purpose:** All features contribute equally to training
   - Notice how the "tall" distribution becomes more circular

**Key insight:** Normalization ensures no single feature dominates due to scale differences.
